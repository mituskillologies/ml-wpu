#### Line Plot

```python
x = np.array([3,12,16,33,45,49,57,71,73,80])
y = np.array([7,11,19,24,21,45,56,78,81,85])
z = np.array([73,71,56,54,44,34,32,26,21,8])
p = np.array([55,57,67,72,55,53,45,49,76,45])
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, z)
```

```python
plt.plot(x, p)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xticks(range(0,91,5))
plt.yticks(range(0,91,5))
plt.xlim(0, 90)
plt.ylim(0, 90)
plt.plot(x, y)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xticks(range(0,91,5))
plt.yticks(range(0,91,5))
plt.xlim(0, 90)
plt.ylim(0, 90)
plt.plot(x, y, label = '2022')
plt.plot(x, z, label = '2023')
plt.legend()
```

```python
plt.figure(figsize=(16,9))
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
# marker => .  o  >  <  v  ^  *  D  + 
plt.plot(x, y, color='brown', marker = 'o')
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
plt.plot(x, y)
plt.subplot(2,1,2)
plt.plot(x, z)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.plot(x, y)
plt.subplot(1,2,2)
plt.plot(x, z)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.plot(x, y, color = 'y')
plt.subplot(1,2,2)
plt.plot(x, z, color = 'r')
plt.subplot(2,2,3)
plt.plot(x, p, color = 'm')
```

#### Bar plot

```python
df = pd.read_csv('student.csv')
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.bar(df['name'], df['marks'], color = 'purple')
```

```python
df['age'].value_counts()
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.xlabel('Age')
plt.ylabel('Count')
plt.bar(df['age'].value_counts().index, df['age'].value_counts(), color = 'gray')
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.ylabel('Student Name')
plt.xlabel('Marks')
plt.barh(df['name'], df['marks'], color = 'navy')
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
plt.grid()
plt.xlabel('Purchased')
plt.ylabel('Count')
plt.bar(df['Purchased'].value_counts().index, df['Purchased'].value_counts(), color = 'g')
```
